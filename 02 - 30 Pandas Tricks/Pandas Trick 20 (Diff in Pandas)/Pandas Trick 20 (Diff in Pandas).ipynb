### Generate Random Data

```python
def generate_sample_data(): # creates a fake df for testing
    number_or_rows = 25
    num_cols = 6
    cols = list("ABCDEF")
    df = pd.DataFrame(np.random.randint(1, 25, size = (number_or_rows, num_cols)), columns=cols)
    df.index = pd.util.testing.makeIntIndex(number_or_rows)
    return df
```

```python
df = generate_sample_data()
df
```

```python
df["A_diff"] = df["A"].diff() # calculate the difference between 2 rows
df
```

```python
 # calculates the porcentual variation between 2 rows
df["A_Diff_Percent"] = df["A"].pct_change()*100
df
```
